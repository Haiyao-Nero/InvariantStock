```python

import pandas as pd

def get_data(stock_list):
    df_all = pd.DataFrame()
    for i in stock_list:
        df = pd.read_sql(f"select * from `{i[0]}`",con=engine)
        df_all = pd.concat([df_all,df])
    df_all.rename(columns={"trade_date":"datetime","ts_code_x":"instrument"},inplace=True)
    df_all.set_index(["datetime","instrument"],inplace=True)
    df_all.drop(columns=["index"],inplace=True)
    return df_all
```

```python
import multiprocessing
num_processes = 16
chunk_size = int(len(stock_list) / num_processes)
chunks = [(stock_list[i:i + chunk_size]) for i in range(0, len(stock_list), chunk_size)]
pool = multiprocessing.Pool(processes=num_processes)
result = pd.concat(pool.map(get_data, chunks))
pool.close()
pool.join()
```
